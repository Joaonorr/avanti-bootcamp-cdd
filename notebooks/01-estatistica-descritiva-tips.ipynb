```python
# Importanto as bibliotecas necessárias
import pandas as pd
import seaborn as sns
```

```python
# carregando o dataframe
df = sns.load_dataset('tips')
df.head(5)
```

```python
# criando o dicionário de dados
df_dict = pd.DataFrame([
    {
        "variavel" : "total_bill",
        "descricao": "valor da conta total",
        "tipo": "quantitativa",
        "subtipo" : "discreta"
    },
    {
        "variavel" : "tip",
        "descricao": "valor da gorjeta",
        "tipo": "quantitativa",
        "subtipo" : "discreta"
    },
    {
        "variavel" : "sex",
        "descricao": "sexo do cliente (male e female)",
        "tipo": "qualitativa",
        "subtipo" : "nominal"
    },
    {
        "variavel" : "smoker",
        "descricao": "se é fumante ou não",
        "tipo": "qualitativa",
        "subtipo" : "nominal"
    },
    {
        "variavel" : "day",
        "descricao": "dia da semana (Sun, Sat, Thur, Fri)",
        "tipo": "qualitativa",
        "subtipo" : "ordinal"
    },
    {
        "variavel" : "time",
        "descricao": "qual refeição (Dinner e Lunch)",
        "tipo": "qualitativa",
        "subtipo" : "nominal"
    },
    {
        "variavel" : "size",
        "descricao": "número de pratos",
        "tipo": "quantitativa",
        "subtipo" : "discreta"
    }
])
df_dict
```

```python
# buscando todas as variáveis qualitativas
qualVet = df_dict.query('tipo == "qualitativa"').variavel.tolist()

# exibindo a moda
df[qualVet].mode()
```

```python
# buscando todas as variáveis quantitativas
quantVet = df_dict.query('tipo == "quantitativa"').variavel.tolist()
df[quantVet].mode()
```

```python
# exibindo a média das variáveis
df[quantVet].mean()
```

```python
# exibindo a mediana das variáveis 
df[quantVet].median()
```

```python
# Exibindo os valores de desvio padrão, invervalo e intervalo interquantil
df.describe()
```

```python
# desvio padrão
df[quantVet].std()
```

```python
# intervalo mínimo
df[quantVet].min()
```

```python
# intervalo máximo
df[quantVet].max()
```

```python
qualVet = df_dict.query('tipo == "qualitativa"').variavel.tolist()
for variavel in qualVet:
    for values in df[variavel].unique():
        print(f'{variavel} - {values}')
        subamostra = df[df[variavel] == values]
        display(subamostra.describe())
        print()
```
